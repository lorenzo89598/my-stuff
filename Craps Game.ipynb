# **Craps Game**

We simulate the Craps game: the shooter (the one who rolls the dice) chooses between a "pass line" or a "don't pass line" bet.



raccoon that gambles.jpg

```python
import random
```

```python
# The number of trials to estimate the probability of winning
trials = 10000
```

#### **Pass Line**



The player wins if the first roll is a "natural" (7 or 11) and loses if it is "craps" (2, 3, and 12). If some other number is rolled, that number become the "point" and the shooter continues to roll. If he rolls the point before rolling a 7, he wins; otherwise, he loses.

```python
# Define the "pass line" unction
def pass_line(num_trials):
    num_wins = 0

    for i in range(num_trials):
        die1 = random.randint(1, 6)
        die2 = random.randint(1, 6)
        sum_dice = die1 + die2

        if sum_dice == 7 or sum_dice == 11:
            num_wins += 1
        elif sum_dice in [2, 3, 12]:
            pass
        else:
            point = sum_dice
            while True:
                die1 = random.randint(1, 6)
                die2 = random.randint(1, 6)
                number = die1 + die2

                if number == point:
                    num_wins += 1
                    break
                if number == 7:
                    break

    return num_wins / num_trials * 100
```

```python
pass_line(trials)
```


####**Don't Pass Line**



The shooter wins if he rolls a 2 or 3 and he loses if 7 or 11 are rolled. If he rolls a 12, it is a push. If some other number is rolled, this becomes the point and the player wins if he rolls a 7 before the point; otherwise, he loses.

```python
# Define the "don't pass line" bet function
def dont_pass_line(num_trials):
    num_wins = 0

    count = 0
    while count < num_trials:
        count += 1
        die1 = random.randint(1, 6)
        die2 = random.randint(1, 6)
        sum_dice = die1 + die2

        if sum_dice == 2 or sum_dice == 3:
            num_wins += 1
        elif sum_dice == 12:
            num_trials += 1
        elif sum_dice in [7, 11]:
            pass
        else:
            point = sum_dice
            while True:
                die1 = random.randint(1, 6)
                die2 = random.randint(1, 6)
                number = die1 + die2

                if number == 7:
                    num_wins += 1
                    break
                if number == point:
                    break

    return num_wins / num_trials * 100
```

```python
dont_pass_line(trials)
```

####**Which choice is better?**

```python
# Percentage of times

def winner_choice(trials):
    pl = pass_line(trials)
    dpl = dont_pass_line(trials)
    if pl > dpl:
        print(f"The winning choice is 'pass line' with a winning probability of {pl}. The difference in winning probability is {pl - dpl}")
    else:
        print(f"The winning choice is 'don't pass line' with a winning probability of {dpl}. The difference in winning probability is {dpl - pl}")

```

```python
winner_choice(trials)
```
